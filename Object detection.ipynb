```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from cvlib.object_detection import draw_bbox
import cvlib as cv
```

```python
def object_detect(string :str):
    image = cv2.imread(string)
    box, label, count = cv.detect_common_objects(image)
    output = draw_bbox(image, box, label, count)
    plt.figure(figsize=(10, 8))
    plt.imshow(output)
    plt.show()
    print("Number of objcts in this image are " +str(label.count('car')))
```

```python
object_detect('cars.jfif')
```

```python
object_detect('images/sheep.jfif')
```

```python
object_detect('images/images.jfif')
```

```python
object_detect('images/traffic.jfif')
```
